```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv('./data/developer_survey_2019/survey_results_public.csv')
```

```python
df.columns
```

```python
def filter_fn(row):
    return ';' not in str(row['Ethnicity'])

single_ethnicity = df[df.apply(filter_fn, axis=1)]
single_ethnicity[['Country', 'Ethnicity']].head()
```

```python
print('Original:', len(df))
print('Single Ethnicity:', len(single_ethnicity))
print('Ratio:', len(single_ethnicity)/len(df))
```

```python
single_ethnicity_living_in_us = single_ethnicity[single_ethnicity['Country'] == 'United States']
living_in_us = df[df['Country'] == 'United States']
print('Single Ethnicity living in USA:', len(single_ethnicity_living_in_us))
print('Living in USA:', len(living_in_us))
print('Ratio:', len(single_ethnicity_living_in_us) / len(living_in_us))
```

```python
relevant_cols = single_ethnicity_living_in_us[['Ethnicity', 'YearsCodePro', 'JobSat', 'ConvertedComp']].dropna(how='any')
relevant_cols = relevant_cols[relevant_cols['YearsCodePro'] != 'Less than 1 year']
relevant_cols = relevant_cols[relevant_cols['YearsCodePro'] != 'More than 50 years']
relevant_cols.head()
```

```python
sat_mapper_lut = {
    'Very dissatisfied' : 0,
    'Slightly dissatisfied' : 0.25,
    'Neither satisfied nor dissatisfied' : 0.5,
    'Slightly satisfied' : 0.75,
    'Very satisfied' : 1,
}
relevant_cols['JobSatNum'] = relevant_cols.apply(lambda row: sat_mapper_lut[row['JobSat']], axis=1)
relevant_cols.head()
```

```python
relevant_cols['YearsCodePro'] = pd.to_numeric(relevant_cols['YearsCodePro'])
relevant_cols.head()
```

```python
relevant_cols['YearsCodeProBucketed'] = relevant_cols.apply(lambda row: (row['YearsCodePro'] // 5) * 5, axis=1)
relevant_cols.head()
```

```python
# INDEX_KEY = 'YearsCodePro'
INDEX_KEY = 'YearsCodeProBucketed'
MIN_GROUP_SIZE = 10
```

```python
grouped_df = relevant_cols.groupby(['Ethnicity', INDEX_KEY])
relevant_cols = relevant_cols[grouped_df['YearsCodeProBucketed'].transform('size') >= MIN_GROUP_SIZE]
print('Sample Size:', len(relevant_cols))

relevant_cols.head()
```

```python
grouped_df = relevant_cols.groupby(['Ethnicity', INDEX_KEY])
```

```python
median_result = grouped_df.median()
median_result = median_result.reset_index()
median_result.head()
```

```python
result_with_index = median_result.set_index(INDEX_KEY)
grouped_with_index = result_with_index.groupby('Ethnicity')
```

```python
grouped_with_index['ConvertedComp'].plot(legend=True, xlabel='Years coding professionally (bucketed 5 years)', ylabel='Compensation USD', figsize=(20,10))
```

```python
grouped_with_index['JobSatNum'].plot(legend=True, xlabel='Years coding professionally (bucketed 5 years)', ylabel='Job Satisfaction', figsize=(20,10))
```
